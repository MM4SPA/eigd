```python
import pandas as pd

# raw annotations
df = pd.read_json("data/eigd/handball/event_annotations.jsonl", lines=True, orient="records")
df.head(5)
```

```python
# number of events per annotator and sequence
# note, three matches are annotated from multiple experts to measure the "average human performance"
df.groupby(["match_id", "subset_id", "annotator"]).agg({"t_start": len}).rename(columns={"t_start": "num_annotations"})
```

```python
# event distribution per annotator
df.groupby("annotator")["label_flat"].value_counts()
```

```python
# number of events for all sequences (5 matches x 5 shots) without multiple annotations 
df.loc[df.annotator == "A"]["label_flat"].value_counts()
```
